Pandas uses backslash \ to show wrapped lines when output is too wide to fit the screen.

```python
import pandas as pd

data = pd.read_csv("data/gapminder_gdp_oceania.csv")
data
```

Use `index_col` to specify that a columnâ€™s values should be used as row headings.

```python
data = pd.read_csv("data/gapminder_gdp_oceania.csv", index_col="country")
data
```

Use the `DataFrame.info()` method to find out more about a dataframe. **Methods have parentheses**. 

```python
data.info()
```

Retrieve information about the dataframe's columns with `dataframe.columns`. 

```python
data.columns
```

To transpose a dataframe, use `dataframe.T`.

```python
data.T
```

`DataFrame.describe()` gets the summary statistics of only the columns that have numerical data. All other columns are ignored, unless you use the argument `include='all'`.

```python
data.describe()
```

```python
americas = pd.read_csv("data/gapminder_gdp_americas.csv", index_col="country")
americas.describe()
```

We can check out the first five rows of americas by executing `americas.head()` (allowing us to view the head of the DataFrame).

```python
americas.head(3)
```

To check out the last three rows of americas, we would use the command, `americas.tail(n=3)`.

```python
americas.tail(3)
```

To display the last three _columns_ of this data, create a new DataFrame in which rows and columns are switched.

```python
americas_flipped = americas.T
americas_flipped
```

```python
americas_flipped.tail(3)
```

We may prefer to display three columns instead of three rows, so we can flip it back.

```python
americas_flipped.tail(n=3).T    
```
